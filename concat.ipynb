```python
# Importing Image class from PIL module 
from PIL import Image 
import os
import torchvision
import torchvision.transforms as transforms
import torch
import matplotlib.pyplot as plt
import numpy as np
from torchvision.utils import save_image
```

```python

def cropimg(img,savepath):
    # Opens a image in RGB mode 

    im = Image.open(img) 

    # Size of the image in pixels (size of original image) 
    # (This is not mandatory) 
    width, height = im.size 

    # Setting the points for cropped image 
    left = 91
    top = 71
    right = 525
    bottom =504

    # Cropped image of above dimension 
    # (It will not change original image) 
    im1 = im.crop((left, top, right, bottom))
    newsize = (300, 300)
    im1 = im1.resize(newsize)
    #im1.save(savepath)
    
    # Shows the image in image viewer 
    
    transform = transforms.Compose([transforms.PILToTensor()])
    tensor = transform(im1)
    im1.close()
    return tensor
```

```python
for i in range(16):
    basepath=os.path.join('visualize_image\\random',  str(i))
    savebasepath = 'visualize_image\\random\\croped'
    specific='Overlayed Integrated Gradients with SmoothGrad Squared'
    images=[]
    for entry in os.listdir(basepath):
        if specific in entry:
            filepath=os.path.join(basepath, entry)
            savepath=os.path.join(savebasepath, entry)
            cropedimg=cropimg(filepath,savepath)
            images.append(cropedimg)
            print(filepath)
    def imshow(img):
        img = img      # unnormalize
        npimg = img.numpy()
        plt.imshow(np.transpose(npimg, (1, 2, 0)))
        plt.show()
    # print images
    combine=torchvision.utils.make_grid(images,nrow=1,padding=10)
    imshow(combine)

    # define a transform to convert a tensor to PIL image
    transform = transforms.ToPILImage()

    # convert the tensor to PIL image using above transform
    img = transform(combine)

    # display the PIL image
    img.show()
    file='visualize_image\\random\\croped\\'+'concat'+str(i)+'.jpg'
    img.save(file)
```
